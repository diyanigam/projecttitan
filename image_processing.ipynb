```python
import math
import matplotlib.pyplot as plt
from itertools import permutations

def connect_points(points):
    connections = []
    for p1, p2 in permutations(points, 2):
        dist = math.dist(p1, p2)
        connections.append((p1, p2, dist))
    connections.sort(key=lambda c: c[2])
    connected_points = []
    for c in connections:
        if c[0] not in connected_points or c[1] not in connected_points:
            connected_points.append(c[0])
            connected_points.append(c[1])
    connected_points.append(connected_points[0])
    x, y = zip(*connected_points)
    plt.plot(x, y, '-o')
    plt.show()
connect_points(line_points)
```
